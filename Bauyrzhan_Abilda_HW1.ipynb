```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
from scipy.spatial.distance import minkowski, chebyshev, mahalanobis

# Name: Bauyrzhan Abilda
# Github Username: b-abilda
# USC ID: 7773067264

file_path = "data/vertebral_column_data/column_2C.dat"  # Change to column_3C.dat if needed
data = pd.read_csv(file_path, sep=r'\s+', header=None)

# Assigning column names
data.columns = ["Pelvic Incidence", "Pelvic Tilt", "Lumbar Lordosis Angle", "Sacral Slope", "Pelvic Radius", "Grade of Spondylolisthesis", "Class"]

# Converting class labels to numeric
if "column_2C" in file_path:
    data["Class"] = data["Class"].map({"NO": 0, "AB": 1})  # Binary classification
elif "column_3C" in file_path:
    data["Class"] = data["Class"].map({"NO": 0, "SL": 1, "DH": 1})  # Merge SL and DH into class 1

# Removing NaN values (if any)
data.dropna(inplace=True)

# Ensure only 0 and 1 in Class
data = data[data["Class"].isin([0, 1])]

# Checking class distribution
class_counts = data["Class"].value_counts()
print("Class distribution in full dataset:")
print(class_counts)

if len(class_counts) < 2:
    raise ValueError("Dataset contains only one class. Classification is not possible.")

# Exploratory Data Analysis 
plt.figure(figsize=(10, 6))
sns.boxplot(data=data.iloc[:, :-1])
plt.title("Boxplot of Features")
plt.show()

sns.pairplot(data, hue="Class", diag_kind='kde')
plt.show()

# Splitting data into train and test sets
y_test = pd.Series([], dtype=int)
random_seed = 42
test_size = 0.3
attempts = 0

while len(y_test.unique()) < 2 and attempts < 10:
    random_seed += 1
    attempts += 1
    X_train, X_test, y_train, y_test = train_test_split(data.iloc[:, :-1], data.iloc[:, -1], 
                                                        test_size=test_size, stratify=data["Class"], random_state=random_seed)
    print(f"Trying random_state={random_seed} with test_size={test_size} to ensure class balance...")
    
    # If still imbalanced, increase test size
    if len(y_test.unique()) < 2 and test_size < 0.5:
        test_size += 0.05

if len(y_test.unique()) < 2:
    raise ValueError("Could not create a balanced test set. Try a different dataset or adjust class proportions.")

# Checking class balance in test set
print("Class distribution in test set:")
print(y_test.value_counts())

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Learning Curve
train_sizes = np.linspace(0.1, 0.9, 9)
train_scores, test_scores = [], []

for train_size in train_sizes:
    X_partial, _, y_partial, _ = train_test_split(X_train_scaled, y_train, train_size=train_size, random_state=42)
    knn = KNeighborsClassifier(n_neighbors=5, metric='euclidean')
    knn.fit(X_partial, y_partial)
    train_scores.append(knn.score(X_partial, y_partial))
    test_scores.append(knn.score(X_test_scaled, y_test))

plt.plot(train_sizes, train_scores, label="Train Score")
plt.plot(train_sizes, test_scores, label="Test Score")
plt.xlabel("Training Set Size")
plt.ylabel("Accuracy")
plt.legend()
plt.title("Learning Curve")
plt.show()

# Adjusting k values to avoid exceeding training size
max_k = min(len(X_train), 30)
k_values = list(range(max_k, 3, -3))

# KNN classification
def knn_experiment(k_values, metric):
    train_errors, test_errors = [], []
    for k in k_values:
        if k > len(X_train):
            continue
        knn = KNeighborsClassifier(n_neighbors=k, metric=metric)
        knn.fit(X_train_scaled, y_train)
        train_errors.append(1 - knn.score(X_train_scaled, y_train))
        test_errors.append(1 - knn.score(X_test_scaled, y_test))
    return train_errors, test_errors

train_errors, test_errors = knn_experiment(k_values, 'euclidean')

plt.plot(k_values, train_errors, label="Train Error")
plt.plot(k_values, test_errors, label="Test Error")
plt.xlabel("k")
plt.ylabel("Error Rate")
plt.legend()
plt.show()

best_k = k_values[np.argmin(test_errors)]
best_knn = KNeighborsClassifier(n_neighbors=best_k, metric='euclidean')
best_knn.fit(X_train_scaled, y_train)
y_pred = best_knn.predict(X_test_scaled)

print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
print(f"Accuracy: {accuracy_score(y_test, y_pred)}")
```
